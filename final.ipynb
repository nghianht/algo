# Case 1: Sinh vien ghep vao truong tuong ung

```python
def init_can(N,diem):
  candiate_case_1 = diem.sample(N)
  candiate_case_1 = candiate_case_1.sort_values(['Khoi_A', 'Toan'], ascending = False)
  candiate_case_1 = candiate_case_1.reset_index()
  candiate_case_1.drop(['index'],axis=1, inplace=True)
  candiate_case_1["STT"] = [i for i in range(len(candiate_case_1))]
  candiate_case_1 = candiate_case_1.values.tolist()
  return candiate_case_1
```

```python
total = diemchuan["chi_tieu"].sum()
nguyen_vong = [["HUST", "UET","PTIT","HUS","HAUI", "UTC","HUCE","TLU","HUNRE"]]*int(total)
```

```python
candiate_case_1 = init_can(len(nguyen_vong),diem)
```

Tao rule Nguyen Vong: NV phia truoc phai la truong co diem chuan cao hon cac NV phia sau, lay lich su cac nam truoc

```python
uni_info = diemchuan.values.tolist()
uni_info
```

```python
#tạo dictionary chứa tên trường và các sinh viên ghép vào trường đó
def create_partner(uni_info):
  uni_partner = {}
  for i in range(len(uni_info)):
    uni_partner[i] = []  
  return uni_partner      
```

```python
def stableMatching(candiate_case_1, nguyen_vong, uni_info):  
  N = len(candiate_case_1)
  uni_partner = create_partner(uni_info)    
  #Khởi tạo hàm kiểm tra sinh viên được nhận hay chưa
  canMatched = [False for i in range(len(candiate_case_1))]  
  #Tạo hàm đếm số sinh viên chưa được nhận
  freeCount = len(candiate_case_1)
  #Tạo vòng lặp khi sinh viên còn chưa được nhận học
  
  while freeCount > 0:
    #Khởi tạo sinh viên đầu tiên
    can = 0
    #Kiểm tra xem sinh viên đó đã có cặp hay chưa, nếu chưa có thì thực hiện ghép căp, nếu đã có cặp thì xét tiếp sinh viên sau
    while(can < N):
      if(canMatched[can] == False):
        break;
      can += 1

    #Tạo vòng lặp với i là thứ tự nguyện vọng của sinh viên can
    i = 0
    #Lần lượt duyệt tất cả các trường của sinh viên can, với can chưa được ghép đôi
    while i < len(nguyen_vong[0]) and canMatched[can]== False:
      #lấy nguyện vọng thứ i của sv can
      uni = nguyen_vong[can][i]
      #tìm số chỉ tiêu của trường ứng với nv i của sv can
      for j in range(len(uni_info)):        
        if uni_info[j][0] == uni:          
          #nếu trường vẫn còn nguyện vọng: thực hiện ghép cặp
          if uni_info[j][1] > 0:               
            uni_partner[j].append(candiate_case_1[can])
            uni_info[j][1] -= 1
            canMatched[can] = True
            freeCount -= 1                  
          else:                   
            if candiate_case_1[can][5] > uni_partner[j][-1][5]:            
              stt = int(uni_partner[j][-1][6])              
              del uni_partner[j][-1]
              uni_partner[j].append(candiate_case_1[can])
              canMatched[stt] = False
              canMatched[can] = True           
      i += 1  
  return uni_partner
            
```

```python
uni_partner = stableMatching(candiate_case_1, nguyen_vong, uni_info)
```
