# Case 1: Sinh vien ghep vao truong tuong ung

```python
diem_abv_25 = diem[diem['Khoi_A'] > 24]
candiate_case_1 = diem_abv_25.sample(n=170)
candiate_case_1 = candiate_case_1.sort_values(['Khoi_A', 'Toan'], ascending = False)
candiate_case_1 = candiate_case_1.reset_index()
candiate_case_1.drop(['index'],axis=1, inplace=True)
```

```python
def create_nv(n,uni):
  nv = []
  for i in range(n):
    randomlist = random.sample(range(len(uni)), 3)
    nv.append(randomlist)
  return nv
```

```python
nv = create_nv(170,uni)
```

```python
candiate_case_1["STT"] = [i for i in range(len(candiate_case_1))]
candiate_case_1
```

Tao rule Nguyen Vong: NV phia truoc phai la truong co diem chuan cao hon cac NV phia sau, lay lich su cac nam truoc

```python
candiate_case_1 = candiate_case_1.values.tolist()
```

```python
uni_info = uni.values.tolist()
```

```python
#tạo dictionary chứa tên trường và các sinh viên ghép vào trường đó
def create_partner(uni_info):
  uni_partner = {}
  for i in range(len(uni_info)):
    uni_partner[i] = []  
  return uni_partner      
```

```python
uni_info
```

```python
def stableMatching(candiate_case_1, nv, uni_info):  
    N = len(candiate_case_1)
    uni_partner = create_partner(uni_info)    
    #Khởi tạo hàm kiểm tra sinh viên được nhận hay chưa
    canMatched = [False for i in range(len(candiate_case_1))]  
    uni_info_cp = uni_info.copy()
    #Tạo vòng lặp để xét tất cả nguyện vọng của thí sinh
    stt_nv = 0
    while stt_nv < len(nv[0]):
                
        #Kiểm tra xem sv đó có ghép cặp hay chưa, nếu chưa thì thực hiện ghép căp, nếu đã có cặp thì xét tiếp sv sau
        for can in range(N):
            if(canMatched[can] == False):                
                #lấy số thứ tự trường là nv thứ i của sinh viên can      
                stt_uni = nv[can][stt_nv]
                #kiểm tra xem trường đó còn chỉ tiêu k, nếu trường đó còn chỉ tiêu thì tiến hành ghép cặp
                if candiate_case_1[can][5]  > uni_info_cp[stt_uni][2]:
                    if uni_info_cp[stt_uni][3] > 0:
                        uni_partner[stt_uni].append(candiate_case_1[can])
                        uni_info_cp[stt_uni][-1] -= 1
                        canMatched[can] = True
                    else:
                    #so sánh điểm thi với sinh viên có điểm thi thấp nhất được nhận vào, nếu > tiến hành ghép cặp, nếu không thì bỏ qua
                        sorted(uni_partner[stt_uni],key=lambda l:l[5], reverse=True)
                        if candiate_case_1[can][5] > uni_partner[stt_uni][-1][5]:         
                        
                            stt = int(uni_partner[stt_uni][-1][6])              
                            del uni_partner[stt_uni][-1]
                            uni_partner[stt_uni].append(candiate_case_1[can])
                            canMatched[stt] = False
                            canMatched[can] = True 
        stt_nv +=1
    return uni_partner
```

```python
uni_partner = stableMatching(candiate_case_1, nv, uni_info)
```
